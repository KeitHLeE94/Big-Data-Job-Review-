```python
lens.get(lens['title'] == 'Pulp Fiction (1994)').rating.mean()
```

Pulp Fiction의 평점 평균을 가져왔다.

```python
lens.groupby('title').rating.min().nlargest(100)
```

최저 평점이 높은 순으로 100개의 영화 제목을 가져왔다.

```python
lens.groupby('user_id').rating.count()
```

각 user가 평점을 남긴 영화의 갯수를 가져왔다.

```python
import numpy as np

movie_stats = lens.groupby('title').agg({'rating': [np.size, np.mean]})

atleast_100 = movie_stats['rating']['size'] >= 100
movie_stats[atleast_100].sort_values(by=[('rating', 'mean')], ascending=False)
```

agg: 한 column 내에 여러개의 함수를 적용하고자 할때 사용.

100개 이상의 평점을 받은 영화에 대해 평점 평균이 높은 순으로 정렬하였다.

```python
places_crosstab = pd.pivot_table(data=frame, values='rating', index='userID', columns='placeID')
places_crosstab.head()
```

user가 각 식당에 대해 남긴 평점을 보여주기 위한 pivot table 생성.

추천할때는 melt하면 안된다.(pivot table로 쓸것)

문제점: sparse하다.

```python
places_crosstab.fillna(0)
```

머신러닝을 적용하기 위해 str을 숫자로 바꾼다: NaN은 0으로 대체.

```python
places_crosstab.corr()
```

식당들끼리의 correlation을 구했다.

```python
temp = places_crosstab.T.corr()
temp['U1004'].dropna()
```

temp: user들끼리의 correlation.

U1004의 correlation 정보를 가져왔다: U1071과 가장 연관이 깊다.

```python
Tortas_ratings = places_crosstab[135085]
Tortas_ratings[Tortas_ratings>=0]
```

특정 식당에 매겨진 평점 정보를 가져왔다.

```python
similar_to_Tortas = places_crosstab.corrwith(Tortas_ratings)

corr_Tortas = pd.DataFrame(similar_to_Tortas, columns=['PearsonR'])
corr_Tortas.dropna(inplace=True)
corr_Tortas
```

135085 식당에 대해 Pearson 상관계수법을 적용하여 상관계수를 계산하였다.

```python
Tortas_corr_summary = corr_Tortas.join(rating['rating_count'])
Tortas_corr_summary[Tortas_corr_summary['rating_count']>=10].sort_values('PearsonR', ascending=False)
```

135085 식당의 상관계수 계산 결과 중 평점의 갯수가 10개 이상인 식당에 대해 상관계수가 높은 순으로 정렬.

```python
places_corr_Tortas = pd.DataFrame([135085, 132754, 135045, 135062, 135028, 135042, 135046], index=np.arange(7), columns=['placeID'])
summary = pd.merge(places_corr_Tortas, cuisine, on='placeID')
summary
```
