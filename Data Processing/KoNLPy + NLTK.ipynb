# KoNLPy + NLTK

형태소 분석 -> 구문 분석 -> 의미 분석 -> 담화 분석.
- 형태소 분석: 입력된 문장을 형태소 단위로 분할하고 품사 부착.
- 구문 분석: 주어, 목적어, 서술어 등 구문 단위 분석.
- 의미 분석: 문장이 의미적으로 옳은 문장인지 판단.
- 담화 분석: 대화 흐름상 어떤 의미를 갖는지 판단.

- KoNLPy: 공개된 형태소 분석기를 개인들이 모여 만든 것.
- NLTK: 영어권 형태소 분석 패키지(Natural Language analysis Tool Kit)

### Hannanum

```python
from konlpy.tag import Hannanum

h = Hannanum()
```

```python
text = '20일 서울고법 형사4부(부장판사 김문석) 심리로 열린 박 전 대통령의 특정범죄 가중처벌 등에 관한 법률 위반(뇌물) 등 혐의 항소심 결심 공판에서 검찰은 박 전 대통령에게 징역 30년에 벌금 1185억원을 내려달라고 요청했다.'
```

```python
textSplit = text.split()
candidates = h.analyze(text)
len(textSplit), len(candidates)
```

```python
h.tagset
```

Hannanum이 가진 모든 tag 종류를 보여준다.

```python
for (i, t) in enumerate(textSplit):
    print(t)
    for c in candidates[i]:
        print("  ", c)
    print()
```

각 어절에 대한 분석 결과를 보여준다.

```python
h.pos(text)
```

어절 단위 품사 분석한 결과.

```python
h.nouns(text)
```

명사만 따로 추려낼 수 있다.

```python
nounlist = t.nouns(text)
len(nounlist)
```

```python
global_unique = list()
global_unique_count = list()
```

```python
for row in text.split('\n'):
    term = t.nouns(row)
    unique = set(term)
    global_unique += list(unique)
    global_unique = list(set(global_unique))
    global_unique_count.append(len(global_unique))
    print(len(unique), len(global_unique))
```

```python
from matplotlib import pyplot
```

```python
pyplot.plot(global_unique_count)
```

## Tokenization

```python
import nltk
from nltk.corpus import gutenberg
```

```python
gutenberg.fileids()
```

```python
emma = gutenberg.open('austen-emma.txt').read()
```

```python
from nltk.tokenize import sent_tokenize

tokens = sent_tokenize(emma)
len(tokens)
```

텍스트를 문장 단위로 쪼갰다.

```python
from nltk import word_tokenize

wtokens = word_tokenize(emma)
len(wtokens)
```

텍스트를 단어 단위로 쪼갰다.

```python
from nltk import regexp_tokenize

p = r"([a-zA-Z0-9]+)+"
retokens = regexp_tokenize(tokens[0], p)
retokens
```

문장을 주어진 정규식을 기준으로 쪼갠다.

```python
text = nltk.Text(retokens)
type(text)
```

```python
text = nltk.Text(wtokens)
type(text)
```

```python
text.vocab()
```

tokenize한 문장을 text에 저장하고 text.vocab을 활용하여 각 토큰이 몇번 나왔는지 볼 수 있다.

```python
print(text.plot(10))
```

빈도가 높은 순으로 그래프 출력.

## 정규화 / 연어

```python
stopwords = ["시발", "씨발"]
```

```python
rowString = "시발 련아 ^^"
```

```python
Kkma().morphs(rowString)
```

우선 주어진 문장에서 형태소 분석 실행.

```python
for src in Kkma().morphs(rowString):
    if src not in stopwords:
        print(src)
```

stopwords에 존재하는 단어가 등장하면 생략.

```python
rowString = "시발련아^^"

Kkma().morphs(rowString)
```

띄어쓰기가 안된 문장에 대해 형태소 분석.

띄어쓰기가 되지 않은 경우 ngram를 활용하여 단어를 걸러낸다.

```python
def bigram(source, n=2):
    result = []
    
    for i in range(len(source)):
        result.append(source[i:i+n])
        
    return result
```

```python
bigram("가나다라")
```

문장을 2글자 단위로 자른다.

```python
rowString = "시발련아^^"
for src in Kkma().morphs(rowString):
    isStop = False
    
    for temp in bigram(src):
        if temp in stopwords:
            isStop = True
    
    if not isStop:
        print(src)
```

우선 주어진 문장에 대한 형태소 분석을 실행한 다음 분석 결과를 2글자 단위로 자른 후 stopwords 등장시 출력하지 않는다.

=> 시발련, 아, ^^ 중 stopword가 존재하는 시발련을 제외한 아, ^^가 출력된다.

```python
import re
```

```python
rowString = "시~#!@J%$!K90j발련아^^ 시1발련아^^ 시!발련아^^"

for src in rowString.split():
    print(src)
    src = re.sub("[^ㄱ-ㅎㅏ-ㅣ가-힣]", "", src)
    print(src)
```

사이에 넣은 특수문자를 전부 없앤다.(한글 문자가 아닌 모든 문자는 없애버린다)
