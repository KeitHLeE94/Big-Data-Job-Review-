## Tidy Data 구조로 만드는 기법
1. melt: 분석하기 편한 데이터로 변환.(column이 값으로 들어간다)
2. stack: wide format -> long format. row로 쌓는다.(결과물: series)
3. unstack: stack으로 변환한 데이터를 다시 wide format으로 변환. 값이 column으로 들어간다.
4. pivot: index가 같은 두 row가 있을 경우 한 column에 대한 값이 다를 경우 에러 발생.(aggregation function이 아니기 때문 - pivot table은 aggregation function이 값으로 들어가므로 에러가 발생하지 않는다.)
5. concatenate
6. join

#### Split, Apply, Combine
- Split: 데이터를 쪼개 같은 속성을 가진 데이터끼리 묶어놓는다.
- Apply: 각종 통계함수 적용.
- Combine: 결과를 보여줄 때 전부 합침.

### 예시 - Ebola 데이터

```python
ebola = pd.read_csv('country_timeseries.csv')
ebola
```

```python
ebola.info()
```

문제점: 국가, status가 column으로 존재한다.

```python
ebola_long = ebola.melt(['Date', 'Day'])
ebola_long.head()
```

melt한 뒤에도 status, 국가를 분류해야하기 때문에 추가작업 필요.

=> _를 기준으로 status, country로 나눠지므로 _을 기준으로 split.

```python
variable_split = ebola_long.variable.str.split('_')

status_value = variable_split.str.get(0)
country_value = variable_split.str.get(1)

ebola_long['status'] = status_value
ebola_long['country'] = country_value
```

```python
ebola_long
```

```python
del ebola_long['variable']
```

```python
ebola_long
```
