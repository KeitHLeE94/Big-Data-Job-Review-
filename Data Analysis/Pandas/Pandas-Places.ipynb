## Folium
- 분석 내용을 지도로 보여주기 위해 사용.

```python
import folium
```

```python
m = folium.Map(location=[37.5665, 126.9780], tiles='OpenStreetMap')
m
```

입력된 위도, 경도를 중심으로 지도 표시.

```python
folium.Marker(location=[37.5311, 126.9113], popup='<i>Timberline Lodge</i>').add_to(m)
m
```

location: Marker를 표시할 위도, 경도 지정.

popup: 아무 html 파일이나 태그 입력 가능.

```python
folium.features.Circle(
    radius=100,
    location=[37.5248, 126.9375],
    popup='The Waterfront',
    color='crimson',
    fill=True,
).add_to(m)
m
```

location: 원 표시할 위도, 경도

popup: 클릭시 출력할 내용

```python
folium.features.CircleMarker(
    location=[37.5311, 126.9113],
    radius=50,
    popup='Laurelhurst Park',
    color='#3186cc',
    fill=True,
    fill_color='#3186cc'
).add_to(m)
m
```

CircleMarker, Circle 차이
- CircleMarker: 지도를 확대하면 작아진다.
- Circle: 지도를 확대하면 커진다.

```python
folium.Marker(location=[37.5122579, 127.07190109999999]).add_to(m)
m
```

```python
folium.features.PolygonMarker(
    locations=[[37.5122579, 127.07190109999999], [37.5248, 126.9375], [37.5311, 126.9113], [37.2967, 126.8363]],
    color='#3186cc',
    fill_color='#3186cc',
    fill_opacity='0.5',
    popup='Seoul'
).add_to(m)
m
```

PolygonMarker: 입력한 위도, 경도에 해당하는 지점들을 꼭지점으로 잡아 지도 상에 평면도형을 그린다.

```python
m.add_child(folium.LatLngPopup())
m
```

```python
m.add_child(folium.ClickForMarker(popup='ClickPoint'))
```

add_child: 지도에 여러 속성을 적용한다.

folium.LatLngPopup: 지도를 클릭하면 클릭한 곳의 위도, 경도가 팝업으로 표시된다.

folium.ClickForMarker: 지도를 클릭하면 클릭한 위치에 marker가 생성된다.

```python
import json

folium.RegularPolygonMarker(
    [37.5665, 126.9780],
    fill_color='#132b5e',
    radius=15,
    popup=folium.Popup(max_width=450).add_child(
        folium.Vega(json.load(open('vis1.json')), width=450, height=250)
    )
).add_to(m)
folium.RegularPolygonMarker(
    [37.4665, 126.9780],
    fill_color='#45647d',
    radius=25,
    popup=folium.Popup(max_width=450).add_child(
        folium.Vega(json.load(open('vis2.json')), width=450, height=250)
    )
).add_to(m)

m
```

마름모꼴 아이콘 클릭시 json으로 저장된 내용이 팝업으로 표시됨.

```python
m2 = folium.Map(location=[37.5665, 126.9780])

seoul_geo_json = r'seoul_municipalities_geo_simple.json'
seoul_topo_json = r'seoul_municipalities_topo_simple.json'

folium.GeoJson(seoul_geo_json, name='geojson').add_to(m2)
folium.TopoJson(open(seoul_topo_json), 'objects.seoul_municipalities_geo', name='topojson').add_to(m2)
folium.LayerControl().add_to(m2)

m2
```

지도에 json파일을 입힘.(서울 구별 경계 표시)
