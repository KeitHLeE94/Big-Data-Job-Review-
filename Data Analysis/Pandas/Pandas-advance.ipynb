```python
data.sort_values(['RegionID', 'State'], ascending=False)
```

RegionID를 기준으로 내림차순 정렬, 같은 값일 경우 State를 기준으로 내림차순 정렬.

```python
data.sort_values(['RegionID', 'State'], ascending=[True, False])
```

RegionID를 기준으로 오름차순 정렬, 같은 값일 경우 State를 기준으로 내림차순 정렬.

sort_values 인자
- by: sort할 기준.
    * 리스트로 넣으면 같은 결과에 대해 하위 sort기준 적용 가능.
- axis 
    * 0: 열
    * 1: 행
- ascending
    * True: 오름차순
    * False: 내림차순.
    * 리스트로 넣으면 by의 순서에 맞게 True, False 설정됨.
- inplace: True로 설정하면 데이터 자체를 sort한 결과로 덮어쓴다.(원본 데이터가 바뀜)

```python
data.sort_values(['RegionID', 'State'], ascending=False, axis=1)
```

에러: axis=1일 경우 sort시 비교할 값이 없기 때문.
