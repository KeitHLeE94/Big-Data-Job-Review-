기본 전처리 방법(pandas의 indexing, slicing 방법)(결과가 dataframe일수도 series일수도 있음)
1. column을 가져오는 방식
    * dict 방식
        - df['title']: series 리턴.
        - df[['title']]: dataframe 리턴.
    * attribute 방식
        - df.title: series 리턴.
        - 컬럼명에 공백 있으면 안됨: 공백 제거해줘야 함.
        - 동시에 2개 이상의 column 못가져온다.
    * filter 방식
        - data.filter(items = ['title'])
        - column이 많을 경우 유용함.
2. row를 가져오는 방식
    * index 방식
        - df.loc[0]: series 리턴.
        - df.loc[[0]]: dataframe 리턴.
    * slicing 방식
        - df.loc[1:3]: dataframe 리턴.
        - df.loc[1:3, 'title']: series 리턴(title column만 가져온다)
    * 조건 방식
        - df[df['rating'].isin(['R'])]
3. row, column 둘 다 가져오려면
    * df.loc[:,['title', 'rating]]
4. 범위 중 하나만 가져오는 방식
    * df.loc[1, 'rating']
    * df.at[1, 'rating']
       - at이 loc보다 빠르기때문에 별도로 만들어짐.
       - 하나의 데이터만 가져오는 메소드.
       - df.ax와 비슷(구버전에서는 ax썼음)
    * df.iloc[1, 1]
    * df.iat[1, 1]
       - iloc, iat: row, column을 index값으로 찾는다.(index가 1인 row, index가 1인 column)
    * df.loc[1:3, 'rating'], df.iloc[1:3, 1]
        - df.loc: 3을 포함한다.
        - df.iloc: 3을 제외한다.(list 검색 방식)

특정 값 수정
- df.loc[1, 'rating'] = 'L'
- df.at[1, 'rating'] = 'L'
