```python
user_col = ['user_id', 'age', 'gender', 'occupation', 'zip code']
users = pd.read_csv('~/Downloads/ml-100k/u.user', sep='|', names=user_col)
users
```

```python
item_col = ["movie_id", "title", "release date", "video release date", "IMDb URL", "unknown", "Action", "Adventure", "Animation", "Children's", "Comedy", "Crime", "Documentary", "Drama", "Fantasy", "Film-Noir", "Horror", "Musical", "Mystery", "Romance", "Sci-Fi", "Thriller", "War", "Western"]
movies = pd.read_csv('~/Downloads/ml-100k/u.item', sep='|', encoding='latin-1', names=item_col, usecols=range(5))
movies
```

usecols: 표시할 column의 갯수를 제한할 수 있다.

```python
movie_rating = pd.merge(movies, ratings)
movie_rating
```

movie_id를 기준으로 movies와 ratings 두 데이터프레임을 합쳤다.

```python
lens = pd.merge(movie_rating, users)
lens
```

user_id를 기준으로 movie_rating과 users 두 데이터프레임을 합쳤다.

```python
lens.title.value_counts()[:100]
```

```python
lens.title.value_counts().nlargest(100)
```

rating 수가 많은 순으로 100개의 결과를 가져왔다.

```python
lens.info()
```

```python
lens.groupby('title').rating.mean().nlargest(100)
```

lens dataframe에서 평점이 높은 순으로 100개의 title을 가져왔다.

```python
lens.groupby('title').rating.min().nlargest(100)
```

lens dataframe에서 평점 수가 적은 순으로 100개의 title을 가져왔다.

```python
lens.groupby('user_id').count()
```

각 user가 남긴 평점 수를 가져왔다.

```python
import numpy as np

movie_stats = lens.groupby('title').agg({'rating': [np.size, np.mean]})

atleast_100 = movie_stats['rating']['size'] >= 100
movie_stats[atleast_100].sort_values(by=[('rating', 'mean')], ascending=False)
```

lens dataframe에서 rating 갯수가 100개 이상인 title을 가져오고 rating 평균이 높은 순으로 정렬하였다.

agg: 하나의 series 내에 2개 이상의 series를 출력하고자 할때 사용.
